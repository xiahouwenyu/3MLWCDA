```python
ra_crab, dec_crab = 83.633083, 22.014500

roi = HealpixConeROI(3.0, 12.0, ra=ra_crab, dec=dec_crab)
maptree = f"/data/home/cwy/Science/3MLWCDA/data/skymapDR_latest/20240731_hsc_out.root"
response = f"/data/home/cwy/Science/3MLWCDA/data/skymapDR_latest/DR_ihep_20240131_hscpsf_mc.root"
# response=f"/data/home/cwy/Science/3MLWCDA/data/Archive/DR_ihep_hscpsf_20240131_new.root"
hawc = HAL("HAWC", 
            maptree, 
            response,
            roi, flat_sky_pixels_size=0.05, n_workers=10)

hawc.set_active_measurements(0, 6)

hawc.display()
```

```python
spectrum = Log_parabola()

source = PointSource("CrabNebula", ra=ra_crab, dec=dec_crab, spectral_shape=spectrum)

# NOTE: if you use units, you have to set up the values for the parameters
# AFTER you create the source, because during creation the function Log_parabola
# gets its units

source.position.ra.bounds = (ra_crab - 0.5, ra_crab + 0.5)
source.position.dec.bounds = (dec_crab - 0.5, dec_crab + 0.5)

spectrum.piv = 10 * u.TeV  # Pivot energy
spectrum.piv.fix = True

spectrum.K = 1e-14 / (u.TeV * u.cm**2 * u.s)  # norm (in 1/(keV cm2 s))
spectrum.K.bounds = (1e-25, 1e-19) # without units energies are in keV

spectrum.beta = 0  # log parabolic beta
spectrum.beta.bounds = (-4., 2.)

spectrum.alpha = -2.5  # log parabolic alpha (index)
spectrum.alpha.bounds = (-4., 2.)

model = Model(source)

data = DataList(hawc)
```

```python
# 定义光谱模型
spectrum = Log_parabola()

# 定义空间模型：高斯分布，延展度为 0.1 deg
spatial_shape = Gaussian_on_sphere(
    lon0=ra_crab,
    lat0=dec_crab,
    sigma=0.01 # * u.deg  # 这里是角度单位
)

# 创建延展源（ExtendedSource）
source = ExtendedSource("CrabNebula", spatial_shape=spatial_shape, spectral_shape=spectrum)

# 设置空间位置边界（用于拟合）
spatial_shape.lon0.bounds = (ra_crab - 0.5, ra_crab + 0.5)
spatial_shape.lat0.bounds = (dec_crab - 0.5, dec_crab + 0.5)

# 设置延展度（sigma）的边界
spatial_shape.sigma.bounds = (0, 5)  # 可以根据实际数据调整范围

# 设置光谱参数
spectrum.piv = 10 * u.TeV  # pivot energy
spectrum.piv.fix = True    # 固定不参与拟合

spectrum.K = 1e-14 / (u.TeV * u.cm**2 * u.s)  # 归一化系数
spectrum.K.bounds = (1e-25, 1e-19)  # 注意：若无单位，默认 keV

spectrum.beta = 0
spectrum.beta.bounds = (-4., 2.)

spectrum.alpha = -2.5
spectrum.alpha.bounds = (-4., 2.)

# 创建模型
model = Model(source)

# 加载数据列表
data = DataList(hawc)
```

```python
ra1, dec1 = ra_crab, dec_crab
#Without extended parameter settings, it's a point source. When you set the corresponding extension, it becomes the respective source.
# crab = my.setsorce("crab",ra1,dec1, fitrange=1,
#                 indexb=(-10,-1), kb=(1e-14my

crab = my.setsorce("crab",ra1,dec1, fitrange=0.5, alpha=-2.6,beta=0.1, sigma=1, sb=(0,3),
                spec=Log_parabola()
                )
# crab2 = my.setsorce("crab2",ra1,dec1, fitrange=0.5, indexb=(-10,-1), kb=(1e-14, 1e-8)
#                 )

ext1 = my.setsorce("ext1",85.5,23.5, fitrange=1, sigma=1, sb=(0,3),
                indexb=(-10,1), kb=(1e-15, 1e-8)
                )

Diffuse = my.set_diffusebkg(
                ra1, dec1,
                # K = 3.3e-13, index = -2.3,
                Kf=False, indexf=False,
                name = "crab"
                )

model = Model(crab, ext1, Diffuse) #

model.display(complete=True)
data = DataList(hawc)
```
